```python
#writes all paper-ids from the DSKG-beat-version into a txt file
import pandas as pd
df = pd.read_csv("/DSKG_BETA.csv")

with open("/MAG_Paper_ID.txt", "w") as outp:
    i = 0
    while i < len(df):
        mag_id = df["isReferencedBy"][i]
        mag_id = mag_id.replace("http://ma-graph.org/entity/", "")
        outp.write(mag_id + "\n")
        i += 1
```

```python
#Writes the IDs of the 19 different disciplines into a text file
field_of_study_set = set()

with open("/paper_disciplines_result.txt") as inp:
    for line in inp:
        paper_id, field_of_study_id = line.split("\t")
        field_of_study_set.add(field_of_study_id)
    
with open("/paper_MAG_Field_Of_Study_ID.txt", "w") as outp:
    for line in field_of_study_set:
        outp.write(line)
```

```python
#writes the results of the adjustment into the beat version of the dskg
import pandas as pd
df = pd.read_csv("/DSKG_BETA.csv", keep_default_na=False)

df['MAG_Disciplines'] = ""


i = 0
while i < len(df.index):
    inp = open("/paper_disciplines_result.txt")
    mag_paper = df['isReferencedBy'][i]
    mag_paper_id = mag_paper.strip("http://ma-graph.org/entity/").strip()
    
    for line in inp:
        paper_id, field_of_study_id = line.split("\t")
        paper_id = paper_id.strip()
        if mag_paper_id == paper_id:
            df['MAG_Disciplines'][i] = str(field_of_study_id) + ""

    i += 1
    #print(i)
    
df.to_csv("/DSKG_BETA_PAPER_FOS.csv")
```

```python
#Write the names of the disciplines in the beat version of the dskg
import pandas as pd
df = pd.read_csv("/DSKG_BETA_PAPER_FOS.csv")

df['MAG_Disciplines_Name'] = ""

i = 0
while i < len(df.index):
    inp = open("/Users/davidlamprecht/Desktop/DSKG_Paper/MAKG_FOS/MAG_Disciplines_labeld.txt")
    mag_discipline = df['MAG_Disciplines'][i]
    mag_discipline = str(mag_discipline).replace(".0", "")
    for line in inp:
        mag_discipline_id, mag_discipline_name = line.split("\t")
        if mag_discipline == mag_discipline_id:
            df['MAG_Disciplines_Name'][i] = mag_discipline_name

    i += 1
    print(i)
    
df.to_csv("/DSKG_BETA_PAPER_FOS_Labeled.csv")
```

```python
#Deletes interfering characters (commas) at the end of the cell entries
import pandas as pd
df = pd.read_csv("/DSKG_BETA.csv")

i = 0
while i < len(df.index):
    
    if str(df["MAG_Disciplines_URI_Percentage"][i]).endswith(", "):
        max_len = len(str(df["MAG_Disciplines_URI_Percentage"][i]))
        unwanted_cut_off = max_len - 2
        df["MAG_Disciplines_URI_Percentage"][i] = str(df["MAG_Disciplines_URI_Percentage"][i])[0:unwanted_cut_off]
        
    if str(df["MAG_Disciplines_Percentage"][i]).endswith(", "):
        max_len = len(str(df["MAG_Disciplines_Percentage"][i]))
        unwanted_cut_off = max_len - 2
        df["MAG_Disciplines_Percentage"][i] = str(df["MAG_Disciplines_Percentage"][i])[0:unwanted_cut_off]
        
    if str(df["MAG_Disciplines_Absolute"][i]).endswith(", "):
        max_len = len(str(df["MAG_Disciplines_Absolute"][i]))
        unwanted_cut_off = max_len - 2
        df["MAG_Disciplines_Absolute"][i] = str(df["MAG_Disciplines_Absolute"][i])[0:unwanted_cut_off]
    i += 1

    
df.to_csv("/DSKG_BETA.csv")  
```
